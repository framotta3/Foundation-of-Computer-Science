### 1. Normalize the location field of *Users* dataset, splitting into city, region, country.

Il dataset *Users* presenta il campo Location in cui sono presenti le informazioni relative alla città, regione e stato. Verranno ora separate rispettivamente nei campi city, region e country.
Il separatore interno al campo Location che permette l'identificazione delle tre informazioni è la virgola.

```python
users[['city','state','country']] = users['Location'].str.split(',', expand=True, n=2)
del users['Location']
users.head()
```

### 2. For each book in the *Books* dataset, compute its average rating.

Il merge dei due dataset è stato fatto tramite il LEFT JOIN in modo da considerare tutti i libri anche quelli per cui non è presente il rating. Ad essi si è associato valore nullo.

```python
book_ratings=pd.merge(books,b_rating,how="left",on='ISBN')
```

Il dataframe presenta quindi una riga per ogni recensione a cui è associato il codice identificativo del libro e dello user che ha rilasciato la recensione.

```python
avg_rating=book_ratings.groupby('ISBN',as_index=False).mean()[['ISBN','Book-Rating']]
avg_rating
```

### 3. For each book in the *GoodBooks* dataset, compute its average rating.

Si noti che nel dataset *Goodbooks* è già presente il rating medio per cui si è confrontato quello calcolato con quello espresso nella campo 'average_rating'.

### 4. Merge together all rows sharing the same book title, author and publisher. We will call the resulting datset `merged books`. The books that have not been merged together will not appear in `merged books`.

Sono stati individuati tutti i duplicati e per ognuno si è salvato un solo record nel dataset *merged_books*.

```python
duplicated=books[books.duplicated(subset =["Book-Title","Book-Author","Publisher"])]
```

```python
merged_books=duplicated.drop_duplicates(subset =["Book-Title","Book-Author","Publisher"])
```

```python
merged_books.head()
```

### 5. For each book in `merged books` compute its average rating.

The average is computed considering all books in `books` that have been merged.

```python
merged_rating=pd.merge(merged_books,book_ratings,on=["Book-Title","Book-Author","Publisher"])
```

```python
avg_merged=merged_rating.groupby(["Book-Title","Book-Author","Publisher"],as_index=False).mean("Book-Rating").rename(columns={'Book-Rating':'Avg'})[["Book-Title","Book-Author","Publisher","Avg"]]
avg_merged.head()
```

### 6. For each book in `merged books` compute the minimum and maximum of the average ratings over all corresponding books in the `books` dataset.

Hence for each book in `merged books` we will have exactly two values (a minimum and a maximum).

```python
minmax_merged=merged_rating.groupby(['Book-Title','Book-Author','Publisher'], as_index=False)['Book-Rating'].agg([min,max])
minmax_merged.head()
```

### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

Dopo aver individuato la lista di autori per ogni libro, è stata calcolato il numero di recensioni per ogni autore.

```python
author_reviews=goodbooks
```

```python
author_reviews['authors']=author_reviews['authors'].str.split(',')
author_reviews['tot_rev']=0
```

```python
author_reviews['authors']
```

```python
for i in range(len(author_reviews)):
    author_reviews.loc[i,['tot_rev']]= author_reviews['work_text_reviews_count'][i]/len(author_reviews['authors'][i])
```

Per ogni libro sono state create tante righe quanti sono gli autori e gli è stato associato il valore di rating tramite la funzione *explode*.

```python
author_reviews=author_reviews.explode('authors')
author_reviews[['authors','original_publication_year','book_id','tot_rev']]
```

Infine, è stato fatto il conteggio del rating per ogni autore.

```python
shared_number_of_reviews_with_a_text=author_reviews.groupby('authors').sum()['tot_rev'].reset_index()
shared_number_of_reviews_with_a_text
```

### 8. For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.

Nel caso in cui un anno presenti più autori che condividono il numero massimo di rating, si è scelto di includerli tutti.

```python
author_year=author_reviews.loc[author_reviews.groupby('original_publication_year')['tot_rev'].idxmax()]
author_year=author_year[['original_publication_year','authors', 'tot_rev']]
author_year
```

### 10. Split the users dataset according to the age. One dataset contains the users with unknown age, one with age 0-14, one with age 15-24, one with age 25-34, and so on.

Si è definita una funzione per l'associazione dell'età ad una classe e si sono poi creati i dataset per classe d'età.

```python
def get_people_count_by_age(df):
    df['age_class'] = pd.cut(x=df['Age'], bins=[0,14,24,34,44,54,64,74,84,94,104], labels=["0_14","15_24","25_34","35_44","45_54","55_64","65_74","75_84","85_94","94_104"])
    return df
```

```python
users_age_class=get_people_count_by_age(users)
users_age_class.head()
```

```python
class_age = users_age_class.groupby('age_class')
bins=10
labels=["0_14","15_24","25_34","35_44","45_54","55_64","65_74","75_84","85_94","94_104"]
df_by_age={}
```

```python
for l in labels:
    df_by_age[l]=class_age.get_group(l)
df_by_age.keys()
```

```python
df_by_age["15_24"].head()
```

Segue la definizione del dataset per i valori nulli nel campo Age.

```python
df_by_age["null_age"]=users[users['age_class'].isnull()]
df_by_age["null_age"].head()
```
